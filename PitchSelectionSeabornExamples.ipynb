```python
import pandas as pd
import seaborn as sns
# Bryant 2018
#df = pd.read_csv("KrisBryantSavant.csv")
#Bryant 2017
df = pd.read_csv("bryant_savant_data.csv")
```

```python
sns.jointplot(x="plate_x", y="plate_z", data=df, size=10)
```

```python
import matplotlib.pyplot as plt
sns.FacetGrid(df, hue="bb_type", size=10, aspect=.5)\
    .map(plt.scatter, "plate_x", "plate_z")\
    .add_legend()
```

```python
sns.boxplot(x="pitch_name", y="launch_speed", data=df)
```

```python
# hard_pitches = df[['pitch_name', 'launch_speed']].copy()
hard_pitches = df.copy()
```

```python
hard_pitches = hard_pitches[(df[['pitch_name']] != "Changeup").all(axis=1)]
hard_pitches = hard_pitches[(df[['pitch_name']] != "Curveball").all(axis=1)]
hard_pitches = hard_pitches[(df[['pitch_name']] != "Knuckle Curve").all(axis=1)]
hard_pitches = hard_pitches[(df[['pitch_name']] != "Slider").all(axis=1)]
hard_pitches = hard_pitches[(df[['pitch_name']] != "Split Finger").all(axis=1)]
hard_pitches = hard_pitches[(df[['pitch_name']] != "Knuckle Ball").all(axis=1)]
hard_pitches = hard_pitches[(df[['pitch_name']] != "Forkball").all(axis=1)]
```

```python
for pitch_name, launch_speed in hard_pitches.groupby('pitch_name'):
    print(launch_speed)
```

```python
ax = sns.boxplot(x="pitch_name", y="launch_speed", data=hard_pitches)
ax = sns.stripplot(x="pitch_name", y="launch_speed", data=hard_pitches, jitter=True, edgecolor="gray")
```

```python
sns.violinplot(x="pitch_name", y="launch_speed", data=hard_pitches, size=10, aspect=2)
```

```python
ax = sns.boxplot(x="zone", y="launch_speed", data=hard_pitches)
# ax = sns.stripplot(x="zone", y="launch_speed", data=hard_pitches, jitter=True, edgecolor="gray")
```

```python
soft_pitches = df[['pitch_name', 'launch_speed']].copy()
soft_pitches = soft_pitches[(df[['pitch_name']] != "2-Seam Fastball").all(axis=1)]
soft_pitches = soft_pitches[(df[['pitch_name']] != "4-Seam Fastball").all(axis=1)]
soft_pitches = soft_pitches[(df[['pitch_name']] != "Sinker").all(axis=1)]
soft_pitches = soft_pitches[(df[['pitch_name']] != "Cutter").all(axis=1)]
soft_pitches = soft_pitches[(df[['pitch_name']] != "Forkball").all(axis=1)]
soft_pitches = soft_pitches[(df[['pitch_name']] != "Knuckle Ball").all(axis=1)]
for pitch_name, launch_speed in hard_pitches.groupby('pitch_name'):
    print(launch_speed)
```

```python
ax = sns.boxplot(x="pitch_name", y="launch_speed", data=soft_pitches)
ax = sns.stripplot(x="pitch_name", y="launch_speed", data=soft_pitches, jitter=True, edgecolor="gray")
```

```python
sns.violinplot(x="pitch_name", y="launch_speed", data=soft_pitches, size=10, aspect=2)
```
